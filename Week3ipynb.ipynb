**F-string :- **

code -


```python
num =  int(input('Enter a number:-'))
for i in range(1,11):
  print(f'{num}*{i} = {num*i}')
```

The 'f' in front od the string diffentiates f-strings from normal strings. f-string is an object which when evaluated rsults in string. The value of the the variable num, i and num*i is inserted in place of {num} {i} {num*i} in the f-string. Two things are important for f-string to do our bidding:
  1. The f in front of the string.
  2. The curly braces enclosing the varaibles.
   

   Code

```python
num = int(input('Enter a number :-'))
for i in range(1,11):
  print('{0}*{1}= {2}'.format(num,i,num*i))
```

There are three pairs of curly braces. The values that go into these three positions are
given as three arguments in the format function. Starting from the left, the first pair of
curly braces in the string is replaced by the first argument in format , the second pair
by the second argument and so on.


The integer inside the curly braces gives the index of the argument in the format
function. The arguments of the format function are indexed from 0 and start from the
left. Changing the order of arguments will change the output.

*Format specifiers:-*

code-


```python
pi  = 22/7
print(f'The value of pi = {pi}')
print('The value of pi = {0}'.format(pi))
print('The value of pi = %.2f'%(pi))
```

Let us look at the content inside the curly braces: {pi:.2f}. The first part before the : is the
variable. Nothing new here. The part after : is called a format specifier. .2f means the
following:
1.  . - this signifies the decimal point.
2.   2 - since this comes after the decimal point, it stipulates that there should be
exactly two numbers after the decimal point. In other words, the value (pi_approx)
should be rounded off to two decimal places.
3.   f - this signifies that we are dealing with a float value.

```python
pi = 22/7
print(f'The value of pi = {pi:.2f}')
print('The value of pi = {0:.2f}'.format(pi))
```

For a better understanding of this concept, let us turn to printing integers with a specific formatting. This time , we will use the format function.

**Break , continue and pass:-**
Break -

```python
#xyz@iitm.in


email = input()
for c in email:
  if c=="@":
    break
  print(c, end= " ")
```

The break statement is used to exit out of a loop without executing any code that comes below it.


**continue**

```python
email = input()
for c in email:
  if c=="@":

    print('')
    continue
  print(c, end = '')
```

The continue statement is used to move to the next iteration of the loop, skipping whatever code comes below it.



Pass -



```python
for x in range (11):
  if x%3 ==0:
    print(x)
  else :
    pass
```

In Python , pass is a null statement. THE interpreter does not ignore a pass statement,
but nothing happens and the statement result in no operation.
