*Generator : A generator is a simple ways to create iterators using a function and yeilds keyword .*

```python
def gen(limit):
    n = 1
    while n <= limit:
        yield n*n   # yield pauses and remembers state
        yield n*n*n
        n += 1


a = gen(3)

print(type(a))


print(next(a))
print(next(a))
print(next(a))
print(next(a))
print(next(a))
print(next(a))
```

*map () : The  map() is used to apply a function to every item in an iterable (list) .*
